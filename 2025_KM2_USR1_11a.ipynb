## Задание 3. Генераторная функция и функция reduce (2 балла)

### Задание 3.1

Напишите генераторную функцию `func3` для генерации нечетных членов арифметической прогрессии $a_1 = 1$, $a_{n+1} =  a_n + 1$, $n=1, 2, \ldots$.

На основе генераторной функции создайте список из 9 первых членов африметической прогрессии.

```python
def gen():
    a=1
    b=1
    while True:
        yield a
        a+=b
l=gen()
for x in range(9):
    print(next(l))
```

Функция-замыкание, которую возвращает фабричная функция, должна увеличивать на один значение переменной `x` из локальной области видимости фабричной функции и возвращать измененное значение в качестве результата с использованием f-строки.

Создайте функцию-замыкание и вызовите ее три раза. Объясните в текстовой ячейке результаты последовательных вызовов функции-замыкания.

```python
def zam(x):
    n=x
    def vn():
        nonlocal n
        n+=1
        return f"{n}"
    return vn
vyzov=zam(8)
for i in range(3):
    print(vyzov())
```
